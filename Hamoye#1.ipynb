<h2><b>Exploratory Data Analysis (EDA) on Fuel Quality Data </b></h2>

<p> This data is from the Federal Energy Regulatory Commission which is provided by the United States Energy Information Administration.</p>

<h6>Importing the necessary libraries</h6>


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

<h6>Importing the dataset</h6>


```python
d_data=pd.read_csv('fuel_ferc1.csv') #giving the dataset the name 'd_data'
d_data
```

```python
#describing the dataset
d_data.describe()
```

```python
#checking out for missing values and duplicate rows
d_data.isnull().any().tolist()
```

<p>The output above implies that the dataset has missing values in one of the columns</p>


```python
#checking for the name of the column that has missing values
d_data.columns[d_data.isnull().any()] .tolist() # 'ms' gves 
```

<p>The column with the name 'fuel_unit' has missing values</p>

```python
#checking out for the amount of fields with missing values
d_data.isnull().sum()
```

<p>Out of 29522 data entries, it is confirmed that only 180 rows have missing values, 180/29522= 0.61%, since this is less than 5% of the whole population, it is better to remove the rows with missing values from the dataset</p>

```python
#droping the rows with the missing values
new_data=d_data.dropna() #new_data is the name of the dataset after having removed the missing values
new_data
```

<p>Now, the dataset has 29,342 data entries</p>

```python
#checking out for any duplicate rows
new_data.duplicated().any()
```

<p>The output indicates that there is no duplicate row</p>

```python
#describing the new dataset
new_data.describe()
```

<h4> Visualizing the dataset</h4>

```python
#plotting the barplots
sns.barplot(data=new_data, x='fuel_type_code_pudl', y='fuel_cost_per_mmbtu')
```

<p>Due to the outliers present, then, it is ok to use the logarithmic scale to make the visualization better </p>

```python
aa=sns.barplot(data=new_data, x='fuel_type_code_pudl', y='fuel_cost_per_mmbtu')
aa.set_yscale("log")
```

<p>The barplot indicates that among all the fuel type codes, <b>waste</b> has the highest fuel cost per mmbtu</p>

```python
#plotting the fuel_type_code_pudl against the fuel cost per unit delivered
sns.barplot(x='fuel_type_code_pudl', y='fuel_cost_per_unit_delivered', data=new_data)
```

<p>Due to the presence of extreme values in the datasets, it is better to use the logarithmic scale, just as the one used earlier</p>

```python
bb=sns.barplot(x='fuel_type_code_pudl', y='fuel_cost_per_unit_delivered', data=new_data)
bb.set_yscale("log")
```

<p>The plot indicates that <b> nuclear</b> has the highest cost per per unit delivered , followed by <b>oil, </b>then <b>coal </b></p>

```python
#plotting fuel type code against fuel cost per unit burned
sns.barplot(x='fuel_type_code_pudl', y='fuel_cost_per_unit_burned', data=new_data)
```

<p>Using the logarithmic scale to better the visualization</p>

```python
cc=sns.barplot(x='fuel_type_code_pudl', y='fuel_cost_per_unit_burned', data=new_data)
cc.set_yscale("log")
```

<p>Just like the previous one done, <b> Nuclear</b> has the highest fuel cost per unit delivered</p>

```python
#plotting fuel mmbtu per unit against fuel type code
sns.barplot(x='fuel_type_code_pudl', y='fuel_mmbtu_per_unit', data=new_data)
```

<p>The plot above indicates that <b>coal</b> has the highest fuel mmbty per unit, followed by <b>nuclear</b></p>

```python
#plotting fuel quantity burned against fuel type code
sns.barplot(x='fuel_type_code_pudl', y='fuel_qty_burned', data=new_data)
```

<p><b>Nuclear</b> has the highest qty burned, then <b>gas</b></p>
